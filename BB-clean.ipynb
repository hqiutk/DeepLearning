## List Comprehension: filter and create new list

the *old way*: find all numbers in a list that is bigger than 5. 
You can paste the follow code to a cell and examine.
```
numbers = list(range(3,9)) # this will create a list [3, 4, 5, 6, 7, 8]
big_numbers = []

for n in numbers:
    if n > 5:
        big_numbers.append(n)

print(big_numbers)
# Output: [5, 7, 8]
```


the *new way*: list comprehension: paste this to cell and examine.

```
# Syntax: [ expression for item in list if condition ]
big_numbers = [n for n in numbers if n > 5]
```

## Exception

The key concept to grasp is your program can fail or crash "unexpectably", and you want to "catch" the error or exception and handle it gracefully. For example (type it out):

```
try:
    # Do the risky thing
    print(10 / 0)
except ZeroDivisionError:
    # Handle the crash gracefully
    print("You can't divide by zero!")
```


## Lambda expression

Standard Function (def): Like buying a heavy-duty tool (like a drill) that you keep in your garage because you use it all the time. You give it a name so you can find it later.

Lambda Expression: Like a "single-use" tool or a quick sticky note. You write it in one line to do one specific math job, use it immediately, and then throw it away. You often don't even give it a name.

The Rule: A lambda can take many arguments, but can only have one expression (one line of logic).


The "Formal" Way (Standard Definition):

```python
def square_it(x):
    return x * x

print(square_it(5))
# Output: 25
```

The "Lambda" Way:

Note: We don't need return. The result is returned automatically.

```python
# syntax: lambda [input]: [logic]
sq = lambda x: x * x

print(sq(5))
# Output: 25
```
